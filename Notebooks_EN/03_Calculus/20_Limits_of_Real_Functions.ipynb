
# **20. Limits of Real Functions**

---

## **1. Definition of Limit**
The limit of a function \( f(x) \) as \( x \) approaches \( a \) is the value \( L \) that \( f(x) \) gets arbitrarily close to as \( x \) gets closer to \( a \):
$$
\lim_{x \to a} f(x) = L
$$
If \( f(x) \) does not approach a finite value, the limit does not exist.

---

## **2. One-Sided Limits**
### **a. Left-Hand Limit**
The left-hand limit of \( f(x) \) as \( x \to a^- \) is:
$$
\lim_{x \to a^-} f(x) = L
$$

### **b. Right-Hand Limit**
The right-hand limit of \( f(x) \) as \( x \to a^+ \) is:
$$
\lim_{x \to a^+} f(x) = L
$$

### **c. Limit Existence**
A limit exists if and only if:
$$
\lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x)
$$

---

## **3. Limits at Infinity**
For \( x \to \infty \) or \( x \to -\infty \), the behavior of \( f(x) \) determines the limit:
$$
\lim_{x \to \infty} f(x) = L \quad \text{or} \quad \lim_{x \to -\infty} f(x) = L
$$

---

## **4. Special Techniques for Limits**
### **a. Direct Substitution**
If \( f(a) \) exists and is finite, then:
$$
\lim_{x \to a} f(x) = f(a)
$$

### **b. Factoring**
Used when direct substitution leads to an indeterminate form like \( \frac{0}{0} \).

### **c. Rationalization**
Used to simplify expressions, especially with square roots.

### **d. L'H么pital's Rule**
If:
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \frac{0}{0} \quad \text{or} \quad \frac{\infty}{\infty},
$$
then:
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}.
$$

$$
\lim_{x \to 2} \frac{x^2 - 4}{x - 2}
$$

$$
\frac{x^2 - 4}{x - 2} = \frac{(x - 2)(x + 2)}{x - 2} = x + 2
$$
So:
$$
\lim_{x \to 2} \frac{x^2 - 4}{x - 2} = 4

### **2. lim (3x^2 + 2x) / (2x^2 + 5) as x -> infinity**

\lim_{x \to \infty} \frac{3x^2 + 2x}{2x^2 + 5}

$$
\frac{3x^2}{2x^2} = \frac{3}{2}
$$

$$
\lim_{x \to \infty} \frac{3x^2 + 2x}{2x^2 + 5} = \frac{3}{2}
