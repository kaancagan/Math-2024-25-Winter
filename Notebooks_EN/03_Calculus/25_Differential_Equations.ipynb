
# **25. Differential Equations**

---

## **1. Definition of Differential Equations**
A differential equation is an equation that relates a function \( y(x) \) to its derivatives.

### **General Form**
$$
F(x, y, y', y'', \dots, y^{(n)}) = 0
$$
where \( y' = \frac{dy}{dx}, \, y'' = \frac{d^2y}{dx^2}, \, \dots \)

---

## **2. Types of Differential Equations**
### **a. Ordinary Differential Equations (ODEs)**
An equation involving ordinary derivatives of \( y(x) \):
$$
\frac{dy}{dx} + P(x)y = Q(x)
$$

### **b. Partial Differential Equations (PDEs)**
An equation involving partial derivatives of \( u(x, t) \):
$$
\frac{\partial u}{\partial t} = D \frac{\partial^2 u}{\partial x^2}
$$

---

## **3. Order and Degree**
### **a. Order**
The order of a differential equation is the highest derivative present.

### **b. Degree**
The degree is the power of the highest derivative, provided the equation is polynomial in derivatives.

---

## **4. Solutions of Differential Equations**
### **a. General Solution**
The solution involving arbitrary constants:
$$
y(x) = C_1f_1(x) + C_2f_2(x) + \dots
$$

### **b. Particular Solution**
The solution obtained by assigning specific values to arbitrary constants.

---

## **5. First-Order Differential Equations**
### **a. Separable Equations**
If:
$$
\frac{dy}{dx} = g(x)h(y),
$$
then:
$$
\int \frac{1}{h(y)} \, dy = \int g(x) \, dx.
$$

### **b. Linear Equations**
If:
$$
\frac{dy}{dx} + P(x)y = Q(x),
$$
then the solution is:
$$
y(x) = e^{-\int P(x) \, dx} \left( \int Q(x)e^{\int P(x) \, dx} \, dx + C \right).
$$

---

## **6. Second-Order Differential Equations**
### **a. Homogeneous Linear Equations**
If:
$$
y'' + P(x)y' + Q(x)y = 0,
$$
then the solution is:
$$
y(x) = C_1y_1(x) + C_2y_2(x),
$$
where \( y_1(x) \) and \( y_2(x) \) are linearly independent solutions.

### **b. Non-Homogeneous Linear Equations**
If:
$$
y'' + P(x)y' + Q(x)y = R(x),
$$
then the solution is:
$$
y(x) = y_h(x) + y_p(x),
$$
where \( y_h(x) \) is the solution to the homogeneous equation, and \( y_p(x) \) is a particular solution.

---

## **7. Special Equations**
### **a. Bernoulli's Equation**
If:
$$
\frac{dy}{dx} + P(x)y = Q(x)y^n,
$$
then:
$$
v = y^{1-n}
$$
reduces it to a linear equation.

### **b. Exact Equations**
If:
$$
M(x, y) + N(x, y)\frac{dy}{dx} = 0,
$$
and:
$$
\frac{\partial M}{\partial y} = \frac{\partial N}{\partial x},
$$
then the solution is:
$$
\int M(x, y) \, dx + \int \left( N(x, y) - \frac{\partial}{\partial y} \int M(x, y) \, dx \right) \, dy = C.
$$

---

## **8. Applications**

   $$
   \frac{dP}{dt} = kP \quad \Rightarrow \quad P(t) = P_0e^{kt}

2. **Newton's Law of Cooling**:

   $$

   $$

```python
from sympy import symbols, Function, dsolve, Eq, Derivative, cos, sin

# TanÄ±mlamalar
x, t = symbols('x t')
y = Function('y')
psi = Function('psi')

# 1. First-order ordinary differential equations
# (a) y'(x) = y
eq1 = Eq(Derivative(y(x), x), y(x))
sol1 = dsolve(eq1)
print(f"1.a) Solution to y'(x) = y: {sol1}")

# (b) y'(x) = 1 / (2y(x))
eq2 = Eq(Derivative(y(x), x), 1 / (2 * y(x)))
sol2 = dsolve(eq2)
print(f"1.b) Solution to y'(x) = 1 / (2y): {sol2}")

# 2. First-order ODEs using separation of variables
# (a) dy/dx = x / y
eq3 = Eq(Derivative(y(x), x), x / y(x))
sol3 = dsolve(eq3)
print(f"2.a) Solution to dy/dx = x / y: {sol3}")

# (b) dy/dx = y / x
eq4 = Eq(Derivative(y(x), x), y(x) / x)
sol4 = dsolve(eq4)
print(f"2.b) Solution to dy/dx = y / x: {sol4}")

# (c) dy/dx = xy
eq5 = Eq(Derivative(y(x), x), x * y(x))
sol5 = dsolve(eq5)
print(f"2.c) Solution to dy/dx = xy: {sol5}")

# 3. Second-order ordinary differential equations
# (a) y''(x) + y'(x) = 0, y(0) = 2, y'(0) = -1
eq6 = Eq(Derivative(y(x), x, 2) + Derivative(y(x), x), 0)
sol6 = dsolve(eq6, ics={y(0): 2, Derivative(y(x), x).subs(x, 0): -1})
print(f"3.a) Solution to y'' + y' = 0: {sol6}")

# (b) y''(x) - y(x) = 0, y(0) = 2, y'(0) = 0
eq7 = Eq(Derivative(y(x), x, 2) - y(x), 0)
sol7 = dsolve(eq7, ics={y(0): 2, Derivative(y(x), x).subs(x, 0): 0})
print(f"3.b) Solution to y'' - y = 0: {sol7}")

# (c) d^2y/dx^2 = -w^2y(x)
w = symbols('w')
eq8 = Eq(Derivative(y(x), x, 2), -w**2 * y(x))
sol8 = dsolve(eq8)
print(f"3.c) Solution to d^2y/dx^2 = -w^2y: {sol8}")

# 4. Check if psi(t, x) satisfies the wave equation
# Wave equation: d^2psi/dt^2 - v^2 * d^2psi/dx^2 = 0
v = symbols('v')
psi_expr = cos(w * t + symbols('k') * x)
wave_eq = Eq(Derivative(psi_expr, t, 2) - v**2 * Derivative(psi_expr, x, 2), 0)
wave_check = wave_eq.simplify()
print(f"4. Wave equation satisfied: {wave_check == Eq(0, 0)}")
```
