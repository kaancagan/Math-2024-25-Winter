
# **24. Integrals**

---

## **1. Definition of the Integral**
### **a. Indefinite Integral**
The indefinite integral (antiderivative) of a function \( f(x) \) is defined as:
$$
\int f(x) \, dx = F(x) + C
$$
where \( F'(x) = f(x) \) and \( C \) is the constant of integration.

### **b. Definite Integral**
The definite integral of \( f(x) \) from \( a \) to \( b \) is:
$$
\int_a^b f(x) \, dx = F(b) - F(a)
$$
where \( F(x) \) is any antiderivative of \( f(x) \).

---

## **2. Properties of Integrals**
1. **Linearity**:
   $$
   \int [af(x) + bg(x)] \, dx = a \int f(x) \, dx + b \int g(x) \, dx
   $$
2. **Zero Integral**:
   $$
   \int_a^a f(x) \, dx = 0
   $$
3. **Reversal of Limits**:
   $$
   \int_a^b f(x) \, dx = -\int_b^a f(x) \, dx
   $$
4. **Additivity**:
   $$
   \int_a^c f(x) \, dx + \int_c^b f(x) \, dx = \int_a^b f(x) \, dx
   $$

---

## **3. Fundamental Theorem of Calculus**
1. **Part 1**:
   If \( F(x) \) is an antiderivative of \( f(x) \), then:
   $$
   \frac{d}{dx} \left( \int_a^x f(t) \, dt \right) = f(x)
   $$
2. **Part 2**:
   If \( f(x) \) is continuous on \( [a, b] \), then:
   $$
   \int_a^b f(x) \, dx = F(b) - F(a)
   $$

---

## **4. Common Integrals**
1. Power Rule:
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1} + C, \quad n \neq -1
   $$
2. Exponential Functions:
   $$
   \int e^x \, dx = e^x + C
   $$
   $$
   \int a^x \, dx = \frac{a^x}{\ln(a)} + C, \quad a > 0, \, a \neq 1
   $$
3. Logarithmic Functions:
   $$
   \int \frac{1}{x} \, dx = \ln|x| + C, \quad x \neq 0
   $$
4. Trigonometric Functions:
   $$
   \int \sin(x) \, dx = -\cos(x) + C
   $$
   $$
   \int \cos(x) \, dx = \sin(x) + C
   $$
   $$
   \int \sec^2(x) \, dx = \tan(x) + C
   $$
5. Inverse Trigonometric Functions:
   $$
   \int \frac{1}{\sqrt{1 - x^2}} \, dx = \arcsin(x) + C, \quad |x| < 1
   $$
   $$
   \int \frac{1}{1 + x^2} \, dx = \arctan(x) + C
   $$

---

## **5. Techniques of Integration**
1. **Substitution**:
   Let \( u = g(x) \), then:
   $$
   \int f(g(x))g'(x) \, dx = \int f(u) \, du
   $$

2. **Integration by Parts**:
   If \( u = u(x) \) and \( v = v(x) \), then:
   $$
   \int u \, dv = uv - \int v \, du
   $$

3. **Partial Fractions**:
   Used to integrate rational functions:
   $$
   \int \frac{P(x)}{Q(x)} \, dx
   $$

---

## **6. Applications of Integrals**
1. **Area Under a Curve**:
   $$
   \text{Area} = \int_a^b f(x) \, dx
   $$
2. **Volume of a Solid**:
   $$
   \text{Volume} = \int_a^b A(x) \, dx
   $$
   where \( A(x) \) is the cross-sectional area.

3. **Arc Length**:
   $$
   L = \int_a^b \sqrt{1 + \left( f'(x) \right)^2} \, dx
   $$

4. **Surface Area**:
   $$
   S = \int_a^b 2\pi f(x) \sqrt{1 + \left( f'(x) \right)^2} \, dx
   $$
