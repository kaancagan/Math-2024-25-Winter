
# **19. Limits of Sequences**

---

## **1. Definition of Limits of Sequences**

A sequence \( \{a_n\} \) is a list of numbers in a specific order:
$$
a_1, a_2, a_3, \dots, a_n, \dots
$$
The limit of a sequence \( \{a_n\} \) as \( n \to \infty \) is the value \( L \) that the terms of the sequence approach:
$$
\lim_{n \to \infty} a_n = L
$$
If no such \( L \) exists, the sequence is said to **diverge**.

---

## **2. Convergence and Divergence**
### **a. Convergence**
A sequence \( \{a_n\} \) converges if there exists a real number \( L \) such that for every \( \epsilon > 0 \), there exists an \( N \) such that for all \( n > N \),
$$
|a_n - L| < \epsilon
$$

### **b. Divergence**
A sequence diverges if it does not converge to a finite value.

### **c. Oscillatory Behavior**
A sequence can oscillate and fail to converge (e.g., \( (-1)^n \)).

---

## **3. Important Limits**
### **a. Basic Limits**
1. $$ \lim_{n \to \infty} \frac{1}{n} = 0 $$
2. $$ \lim_{n \to \infty} r^n = 0 \quad \text{for } |r| < 1 $$
3. $$ \lim_{n \to \infty} n^p = \infty \quad \text{for } p > 0 $$
4. $$ \lim_{n \to \infty} \frac{\ln n}{n} = 0 $$

### **b. Arithmetic Properties**
1. If $$ \lim_{n \to \infty} a_n = A $$ and $$ \lim_{n \to \infty} b_n = B $$, then:
   - $$ \lim_{n \to \infty} (a_n + b_n) = A + B $$
   - $$ \lim_{n \to \infty} (a_n b_n) = AB $$
   - $$ \lim_{n \to \infty} \frac{a_n}{b_n} = \frac{A}{B}, \quad \text{if } B \neq 0 $$

---

## **4. Squeeze Theorem**
If $$ a_n \leq b_n \leq c_n $$ for all $$ n \geq N $$, and
$$
\lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L,
$$
then:
$$
\lim_{n \to \infty} b_n = L
$$

---

## **5. Examples of Common Sequences**
### **a. Geometric Sequence**
For $$ a_n = r^n $$:
- If $$ |r| < 1 $$, then $$ \lim_{n \to \infty} r^n = 0 $$.
- If $$ |r| > 1 $$, the sequence diverges.

### **b. Harmonic Sequence**
$$
a_n = \frac{1}{n} \quad \Rightarrow \quad \lim_{n \to \infty} a_n = 0
$$

### **c. Exponential Sequence**
$$
a_n = \frac{n^k}{e^n}, \quad k > 0 \quad \Rightarrow \quad \lim_{n \to \infty} a_n = 0
$$

---

## **6. Monotonic and Bounded Sequences**
### **a. Monotonic Sequence**
A sequence is monotonic if it is either:
- Increasing: $$ a_{n+1} \geq a_n $$
- Decreasing: $$ a_{n+1} \leq a_n $$

### **b. Bounded Sequence**
A sequence $$ \{a_n\} $$ is bounded if there exist real numbers $$ m $$ and $$ M $$ such that:
$$
m \leq a_n \leq M, \quad \forall n
$$

### **c. Monotone Convergence Theorem**
If a sequence is monotonic and bounded, it converges.

---

## **7. Recursive Sequences**
A recursive sequence is defined by a formula that relates each term to the previous one. For example:
$$
a_{n+1} = f(a_n)
$$
- To find the limit, solve $$ L = f(L) $$.

---

## **8. Infinite Limits**
- If $$ a_n \to \infty $$, we write:
$$
\lim_{n \to \infty} a_n = \infty
$$

- If $$ a_n \to -\infty $$, we write:
$$
\lim_{n \to \infty} a_n = -\infty
$$

---

## **9. Special Techniques**
### **a. Ratio Test for Sequences**
For $$ a_n > 0 $$, if:
$$
\lim_{n \to \infty} \frac{a_{n+1}}{a_n} = L,
$$
- $$ L < 1 $$: Sequence converges.
- $$ L > 1 $$: Sequence diverges.
- $$ L = 1 $$: Test is inconclusive.

### **b. Root Test for Sequences**
For $$ a_n \geq 0 $$, if:
$$
\lim_{n \to \infty} \sqrt[n]{a_n} = L,
$$
- $$ L < 1 $$: Sequence converges.
- $$ L > 1 $$: Sequence diverges.
- $$ L = 1 $$: Test is inconclusive.

---

   $$
   \lim_{n \to \infty} \frac{2n + 3}{5n - 4}
   $$

2. Prove the convergence of the sequence $$ a_n = \frac{1}{n^2} $$.

3. Determine whether the sequence $$ a_n = (-1)^n \frac{1}{n} $$ converges or diverges.

4. Show that the harmonic sequence $$ a_n = \frac{1}{n} $$ converges to $$ 0 $$.

```python
import numpy as np
from sympy import limit, symbols, sin, oo, exp

# 1. Calculate the limits
n = symbols('n')

# (a) lim (n^2 + 3n) / (2n^2 - 2n)
expr1 = (n**2 + 3*n) / (2*n**2 - 2*n)
limit1 = limit(expr1, n, oo)
print(f"1.a) lim (n^2 + 3n) / (2n^2 - 2n) = {limit1}")

# (b) lim ((2n + 3)^3) / (n^3 - 1)
expr2 = ((2*n + 3)**3) / (n**3 - 1)
limit2 = limit(expr2, n, oo)
print(f"1.b) lim ((2n + 3)^3) / (n^3 - 1) = {limit2}")

# 2. Prove using the squeeze theorem: lim (sin(n) / n)
expr3 = sin(n) / n
limit3 = limit(expr3, n, oo)
print(f"2) lim (sin(n) / n) = {limit3}")
# The limit is proven by observing that -1/n <= sin(n)/n <= 1/n and applying the Squeeze Theorem.

# 3. Find the limit of the sequence: a_n = (1 + 1/n)^n
expr4 = (1 + 1/n)**n
limit4 = limit(expr4, n, oo)
print(f"3) lim (1 + 1/n)^n = {limit4}")
# This limit is well-known to converge to Euler's number e.
```
