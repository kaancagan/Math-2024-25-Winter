# **18. Functions**

---

## **1. Definition of Functions**
A function \( f \) is a rule that assigns to each element \( x \) in a set \( D \) (called the domain) exactly one element \( f(x) \) in a set \( R \) (called the range).

- **Notation**:
  $$
  f: D \to R, \quad x \mapsto f(x)
  $$

- **Example**:
  $$
  f(x) = x^2
  $$

---

## **2. Domain and Range**
- **Domain**: The set of all possible input values for which the function is defined.
- **Range**: The set of all possible output values of the function.

- **Example**:
  $$
  f(x) = \frac{1}{x}
  $$
  - Domain: \( x \neq 0 \)
  - Range: \( y \neq 0 \)

---

## **3. Types of Functions**
### **a. Polynomial Functions**
$$
f(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0
$$
- Example: \( f(x) = 2x^3 + 3x^2 - x + 7 \)

### **b. Rational Functions**
$$
f(x) = \frac{P(x)}{Q(x)}
$$
- Example: \( f(x) = \frac{x^2 + 1}{x - 2} \)

### **c. Trigonometric Functions**
- Sine: \( \sin(x) \)
- Cosine: \( \cos(x) \)
- Tangent:
$$
\tan(x) = \frac{\sin(x)}{\cos(x)}
$$

### **d. Exponential Functions**
$$
f(x) = a^x, \quad a > 0, a \neq 1
$$
- Example: \( f(x) = 2^x \)

### **e. Logarithmic Functions**
$$
f(x) = \log_a(x), \quad x > 0, a > 0, a \neq 1
$$
- Example: \( f(x) = \ln(x) \) (natural logarithm)

---

## **4. Composite Functions**
- Combining two functions \( f \) and \( g \) results in a composite function:
$$
(f \circ g)(x) = f(g(x))
$$

- **Example**:
  If \( f(x) = x^2 \) and \( g(x) = x + 3 \),
  $$
  (f \circ g)(x) = f(g(x)) = (x + 3)^2
  $$

---

## **5. Even and Odd Functions**
- **Even Functions**:
$$
f(-x) = f(x), \quad \text{symmetry about the } y\text{-axis}.
$$
  - Example: \( f(x) = x^2 \)

- **Odd Functions**:
$$
f(-x) = -f(x), \quad \text{symmetry about the origin}.
$$
  - Example: \( f(x) = x^3 \)

---

## **6. Inverse Functions**
- A function \( f \) has an inverse \( f^{-1} \) if:
$$
f(f^{-1}(x)) = x \quad \text{and} \quad f^{-1}(f(x)) = x
$$

- **Example**:
  $$
  f(x) = 2x + 3, \quad f^{-1}(x) = \frac{x - 3}{2}
  $$

---

## **7. Piecewise Functions**
Functions defined by different expressions in different intervals.

- **Example**:
$$
f(x) =
\begin{cases}
x^2, & \text{if } x \geq 0, \\
-x, & \text{if } x < 0.
\end{cases}
$$

---

## **8. Graphing Functions**
- Use key points, asymptotes, and behavior at \( x \to \infty \) to sketch graphs.
- Example: For \( f(x) = \frac{1}{x} \), graph has:
  - Vertical asymptote at \( x = 0 \)
  - Horizontal asymptote at \( y = 0 \)

---

## **9. Transformations of Functions**
### **a. Vertical Shifts**
$$
f(x) + c \quad (\text{shift } c \text{ units up if } c > 0)
$$
### **b. Horizontal Shifts**
$$
f(x + c) \quad (\text{shift } c \text{ units left if } c > 0)
$$
### **c. Scaling**
$$
af(x) \quad (\text{vertical stretch if } |a| > 1, \text{ compress if } |a| < 1)
$$
### **d. Reflections**
$$
-f(x) \quad (\text{reflection about the } x\text{-axis})
$$

---

## **10. Special Properties**
- **Injective (One-to-One)**: Each \( y \) value is mapped by a unique \( x \).
- **Surjective (Onto)**: Every \( y \) value in the range is covered.
- **Bijective**: Both injective and surjective.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Draw the functions and calculate their values at x = 2.
x = np.linspace(-10, 10, 500)
f = lambda x: x**2
g = lambda x: np.sqrt(x)
h = lambda x: 1/x
j = lambda x: np.sin(x)

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="f(x) = x^2")
plt.plot(x[x >= 0], g(x[x >= 0]), label="g(x) = sqrt(x)")  # Domain restriction
plt.plot(x[x != 0], h(x[x != 0]), label="h(x) = 1/x")  # Exclude x = 0
plt.plot(x, j(x), label="j(x) = sin(x)")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Functions")
plt.show()

# Calculate values at x = 2
x_val = 2
f_val = f(x_val)
g_val = g(x_val)
h_val = h(x_val)
j_val = j(x_val)

print(f"f(2) = {f_val}")
print(f"g(2) = {g_val}")
print(f"h(2) = {h_val}")
print(f"j(2) = {j_val}")

# 2. Composite functions f(g(x)), g(f(x)), f(f(x)), g(g(x))
f = lambda x: 3 * x - 1
g = lambda x: np.sqrt(x)

f_of_g = f(g(x_val))
g_of_f = g(f(x_val))
f_of_f = f(f(x_val))
g_of_g = g(g(x_val))

print(f"f(g(2)) = {f_of_g}")
print(f"g(f(2)) = {g_of_f}")
print(f"f(f(2)) = {f_of_f}")
print(f"g(g(2)) = {g_of_g}")

# 3. f(x) = e^x, g(x) = ln(x): Check f(g(x)) and g(f(x))
f = lambda x: np.exp(x)
g = lambda x: np.log(x)

x_val = 2  # Example x value
f_of_g = f(g(x_val))
g_of_f = g(f(x_val))

print(f"f(g(x)) at x = {x_val} = {f_of_g}")
print(f"g(f(x)) at x = {x_val} = {g_of_f}")
print("Notice: f(g(x)) = x and g(f(x)) = x for x > 0")

# 4. Inverse function for f = {(1,7), (2,9), (3,11)}
f_pairs = {1: 7, 2: 9, 3: 11}
f_inverse = {v: k for k, v in f_pairs.items()}
print(f"Inverse function f^-1: {f_inverse}")

# 5. Inverse function for f = {(1,7), (2,7), (3,11)}
f_pairs = {1: 7, 2: 7, 3: 11}
# Note: Inverse does not exist for non-injective functions
print("Inverse does not exist as the function is not injective.")

# 6. f(x) = x - 1. Inverse function f^-1 and visualization
f = lambda x: x - 1
f_inverse = lambda y: y + 1

# Visualization
x = np.linspace(-10, 10, 500)
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="f(x) = x - 1")
plt.plot(x, f_inverse(x), label="f^-1(x) = x + 1", linestyle="--")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Function and Its Inverse")
plt.show()
```
